```python
# Month-to-Date (MTD) Funded Amount
# first we have to convert issue_date to datatime 
# Convert 'Issue Date' column to datetime format
data['issue_date'] = pd.to_datetime(data['issue_date'],format='%d-%m-%Y')

# Set the current month (e.g., August 2021)
current_month = pd.to_datetime("01-01-2021")

# Filter data to current month
data_current_month = data[data['issue_date'].dt.to_period('M') == current_month.to_period('M')]

# Calculate MTD Funded Amount
mtd_funded_amount = data_current_month['loan_amount'].sum()

print("MTD Funded Amount:", mtd_funded_amount)
```

```python

# Convert 'Issue Date' column to datetime format
data['issue_date'] = pd.to_datetime(data['issue_date'],format='%d-%m-%Y')

# Calculate MTD Funded Amount for each month
monthly_funded_amount = data.groupby(data['issue_date'].dt.to_period('M'))['loan_amount'].sum()

print(monthly_funded_amount)
a=monthly_funded_amount.sum()
print(a)
```
